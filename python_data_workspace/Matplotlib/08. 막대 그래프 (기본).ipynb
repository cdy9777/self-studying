# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False  # 글씨체를 한글로 바꾸었을 때 -(마이너스) 깨짐 현상 방지
```

```python
labels = ["강백호", "서태웅", "정대만"]
values = [190, 187, 184]

plt.bar(labels, values) # 막대그래프 .bar(라벨, 값)
```

```python
# labels = ["강백호", "서태웅", "정대만"]
# values = [190, 187, 184]
plt.bar(labels, values, color = "r") # 막대그래표 색 지정
```

```python
# labels = ["강백호", "서태웅", "정대만"]
# values = [190, 187, 184]
colors = ["r","g", "b"]
plt.bar(labels, values, color = colors) # 막대별 color도 다르게 표시 가능
```

```python
# labels = ["강백호", "서태웅", "정대만"]
# values = [190, 187, 184]
plt.bar(labels, values, color = colors, alpha = 0.5) # 막대별 color도 다르게 표시 가능
```

```python
# labels = ["강백호", "서태웅", "정대만"]
# values = [190, 187, 184]
colors = ["r","g", "b"]
plt.bar(labels, values, color = colors) # 막대별 color도 다르게 표시 가능
plt.ylim(175,195) # ylim() y축 값 범위 지정
```

```python
plt.bar(labels, values, width = 0.5)
```

```python
plt.bar(labels, values, width = 0.3)
plt.xticks(rotation = 45) # x축 데이터 각도를 45도로 설정
plt.yticks(rotation = 45) # y축 데이터 각도를 45도로 설정
```

```python
# labels = ["강백호", "서태웅", "정대만"]
# values = [190, 187, 184]
ticks = ["1번학생", "2번학생", "3번학생"]

plt.bar(labels, values)
plt.xticks(labels, ticks) # 레이블에 있는 데이터 이름을 바꿀 수 있음.
```
