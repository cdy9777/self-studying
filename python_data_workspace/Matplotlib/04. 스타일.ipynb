# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False  # 글씨체를 한글로 바꾸었을 때 -(마이너스) 깨짐 현상 방지

x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y,linewidth = 5) #선 두께
```

## marker(마커)

```python
plt.plot(x,y, marker= 'o') # 데이터 표시
```

```python
plt.plot(x,y, marker= 'o', linestyle = "None")
```

```python
plt.plot(x,y, marker = 'v', markersize = '10') # 마크 사이즈 조절 가능. matplotlib marker 구글링 시 다양한 marker 나옴.
```

```python
plt.plot(x,y, marker = 'o', markersize = '10', markeredgecolor = "red") # markedgecolor
```

```python
plt.plot(x,y, marker = 'o', markersize = '20', markeredgecolor = "red", markerfacecolor = "yellow") # markfacecolor
```

## 선 스타일

```python
plt.plot(x,y, linestyle = ":")
```

```python
plt.plot(x,y, linestyle = "--")
```

```python
plt.plot(x,y, linestyle = "-.") # matplotlib linestyles 구글링 시 다양한 스타일 parameter 찾기 가능
```

## 색깔

```python
plt.plot(x, y, color = "red")
```

```python
plt.plot(x, y, color = "green") # matplotlib color 구글링 해도 다 나옴.
```

## 포맷

```python
plt.plot(x,y, "ro--") #red + o marker + --
```

```python
plt.plot(x, y, "bv:") # blue + v + :
```

```python
plt.plot(x, y, "go") # green + o + None
```

## 축약어

```python
plt.plot(x,y, color = "green", marker = "o", ms = 10, mec = "blue", mfc = "red", ls = ":") # matplotlib marker mfc 구글링
```

## 투명도

```python
plt.plot(x,y, marker = "o", mfc = "red", ms = 10, alpha = 0.3) # alpha 투명도 (0 ~ 1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5)) # 그래프 크기 조절
plt.plot(x,y)
```

```python
plt.figure(figsize = (10,5), dpi = 100) # dpi = dots per inch , 확대의 개념 
plt.plot(x,y)
```

## 그래프 배경색

```python
plt.figure(facecolor = "blue")
plt.plot(x,y)
```
